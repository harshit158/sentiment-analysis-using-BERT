```python
from fractions import Fraction
def p(event, space):
    if callable(such_that):
        event=such_that(event, space)
    return Fraction(len(event & space), len(space))
def such_that(predicate, collection):
    return collection.intersection(predicate)
    return {e for e in collection if predicate(e)}
    
```

```python
from ipywidgets import interact, widgets
from IPython.display import display, clear_output

def plot(words, scores):
    plt.xticks(fontsize=14, rotation=45)
    plt.bar(words, scores)

def get_widget1():
    text_wdgt = widgets.Text(
        value='',
        placeholder='Paste ticket description here!',
        description='String:',
        disabled=False
    )
    display(text_wdgt)
    
    output = widgets.Output()
    display(output)
    
    scores = []
    words = []

    def callback(text_wdgt):
        with output:
            new_text = text_wdgt['new']
            if new_text and new_text[-1]==' ':
                clear_output()
                out=p.predict(new_text)
                scores.append(out)
                words.append(new_text.split()[-1])
                print(new_text)
                plot(words, scores)
            clear_output()
    
    return text_wdgt.observe(handler=callback, names='value')
```
